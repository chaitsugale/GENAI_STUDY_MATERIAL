```python
from transformers import AutoTokenizer
import transformers
import torch
```

```python
#model_id = "chaitz2727/Llama-2-7b-chat-new-finetune"
model_id = "snshrivas10/Llama-2-7b-chat-new-finetune"
```

```python
prompt="what is a largest language model?"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_id)
```

```python
model = transformers.pipeline(
  "text-generation",
    model=model_id,
    torch_dtype=torch.float16,
    device_map="auto",
)
```

```python
f"<s>[INST] {prompt} [/INST]",
```

```python
tokenizer.eos_token_id
```

```python
sequences = model(
    f'<s>[INST] {prompt} [/INST]',
    do_sample=True,
    top_k=3,
    num_return_sequences=1,
    eos_token_id=tokenizer.eos_token_id,
    max_length=200,
)
```

```python
sequences[0]['generated_text']
```
